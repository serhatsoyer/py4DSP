### Band-Stop FIR Filter

Next example: [/examples/lti_filters/fir/all-pass.ipynb](https://github.com/serhatsoyer/py4DSP/blob/main/examples/lti_filters/fir/all-pass.ipynb)

```python
import sys
sys.path.insert(0, '../../../') # To be able to reach functions
from functions import fft, lti
from scipy import signal
```

```python
num_of_coeffs = 81 # Number of coefficients, i.e. the filter order + 1
cutoff_freqs = [.2, .6]
numerator_coeffs = signal.firwin(num_of_coeffs, cutoff_freqs) # FIR (LPF by default)
lti.equation(numerator_coeffs)
```

```python
lti.plot_filter_coefficients(numerator_coeffs)
lti.plot_impulse_and_step_responses(numerator_coeffs)
lti.bode(numerator_coeffs)
```

```python
first_time = 0 # s
last_time = 3 # s
samp_freq = 100 # Hz

time = fft.get_time(first_time, last_time, samp_freq)

# cutoff_freqs = [.2, .6]
# 1 means 50 Hz
# .2 means 10 Hz
# .6 means 30 Hz
# Middle of the band: 20 Hz
# Completely pass: 5 Hz
# Completely pass: 35 Hz

zero_freq = 0 # Hz
low_freq = 5 # Hz
low_cutoff_freq = 10 # Hz
middle_band_freq = 20 # Hz
high_cutoff_freq = 30 # Hz
high_freq = 35 # Hz
very_high_freq = 49.5 # Hz

common_amplitude = 1
common_phase = 0 # rad
common_offset = 0

signal_zero_freq = fft.get_signal(zero_freq, 0, common_phase, common_amplitude, time)
signal_low_freq = fft.get_signal(low_freq, common_amplitude, common_phase, common_offset, time)
signal_low_cutoff_freq = fft.get_signal(low_cutoff_freq, common_amplitude, common_phase, common_offset, time)
signal_middle_band_freq = fft.get_signal(middle_band_freq, common_amplitude, common_phase, common_offset, time)
signal_high_cutoff_freq = fft.get_signal(high_cutoff_freq, common_amplitude, common_phase, common_offset, time)
signal_high_freq = fft.get_signal(high_freq, common_amplitude, common_phase, common_offset, time)
signal_very_high_freq = fft.get_signal(very_high_freq, common_amplitude, common_phase, common_offset, time)

superposition_signal = signal_zero_freq + signal_low_freq + signal_high_freq + signal_very_high_freq
superposition_signal += signal_low_cutoff_freq + signal_middle_band_freq + signal_high_cutoff_freq
```

```python
fft.plot_signal_and_fft(time, superposition_signal, samp_freq)
filtered_signal = signal.lfilter(numerator_coeffs, [1], superposition_signal)
fft.plot_signal_and_fft(time, filtered_signal, samp_freq)
```
