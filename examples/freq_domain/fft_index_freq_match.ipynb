### Imperfect FFT Index - Frequency Match

- When the condition $f=\frac{k}{N}$ (or $k=fN$) cannot be satisfied for an $f$ with any $k$
- We have signal amplitude - FFT amplitude mismatch (poor frequency resolution)
- Still, FFT well approximates the frequencies in the example

Next example: [/examples/freq_domain/spectrogram.ipynb](https://github.com/serhatsoyer/py4DSP/blob/main/examples/freq_domain/spectrogram.ipynb)
