### Plotting Spectrogram Using *fft.py*

Next example: [/examples/lti_filters/filter_bank.ipynb](https://github.com/serhatsoyer/py4DSP/blob/main/examples/lti_filters/filter_bank.ipynb)

```python
import sys
sys.path.insert(0, '../../') # To be able to reach functions
from functions import fft
from scipy.signal import chirp
```

```python
# Settings Cell
first_time = 0 # s
last_time = 30 # s
samp_freq = 100 # Hz
first_freq_ver1 = 0 # Hz
first_freq_ver2 = 1 # Hz
last_freq = (samp_freq / 2) - 1 # Hz
```

```python
time = fft.get_time(first_time, last_time, samp_freq)
```

```python
signal = chirp(time, f0=first_freq_ver1, f1=last_freq, t1=time[-1], method='linear')
fft.plot_signal_and_spectrogram(time, signal, samp_freq)
```

```python
signal = chirp(time, f0=first_freq_ver1, f1=last_freq, t1=time[-1], method='quadratic')
fft.plot_signal_and_spectrogram(time, signal, samp_freq)
```

```python
signal = chirp(time, f0=first_freq_ver2, f1=last_freq, t1=time[-1], method='logarithmic')
fft.plot_signal_and_spectrogram(time, signal, samp_freq)
```

```python
signal = chirp(time, f0=first_freq_ver2, f1=last_freq, t1=time[-1], method='hyperbolic')
fft.plot_spectrogram(time, signal, samp_freq) # To show that it is also possible to use
fft.plot_fft(signal, samp_freq)
```
